### The following will demonstrate, with a few examples, how to use our model to compute the condition number in transcript space.

```python
from ComputeC import computeC
from Process_gene_data import process_gene_data
import numpy as np
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes
from matplotlib.ticker import FixedLocator
from matplotlib.lines import Line2D
```

#### Example 1

```python
# Example 1
filename = 'Gene_Data/example_1.txt'
arrays, gene_dict = process_gene_data(filename)
print(arrays)
print(gene_dict)
```

```python
# Generate the range of read lengths
readlengh_range = np.linspace(100, 2501, 25, endpoint=True, dtype=int)

# Initialize the condition number list
condition_number = []

# Compute the condition number for each read length
for readlengh in readlengh_range:
    C = computeC(arrays, readlengh, gene_dict)
    eigenvalues, eigenvectors = np.linalg.eig(C.T @ C)
    condition_number_temp = max(eigenvalues) / min(eigenvalues)
    condition_number.append(condition_number_temp)
```

```python
# Create the main figure
fig = plt.figure(figsize=(15, 12), dpi=600)

# Create a broken y-axis plot
bax = brokenaxes(ylims=((0, 5), (11965, 12000)), hspace=.2, d=0.005)  # Adjust ylims and hspace to fit your data

# Plot the data
line1 = bax.plot(readlengh_range, condition_number, 'o-', markersize=10, color='crimson', label='Condition Number', linewidth=4)

# Set axis labels
bax.set_xlabel('Read Length', fontsize=45, labelpad=45)
bax.set_ylabel('Condition Number', fontsize=45, labelpad=115)

# Create proxy artist
proxy = [Line2D([0], [0], color='crimson', marker='o', linestyle='-', markersize=5)]

# Add grid lines
bax.grid(True, linestyle='--', alpha=1.0, linewidth=2.0)

# Set the positions of the vertical grid lines
grid_positions = [100, 500, 900, 1300, 1700, 2100, 2500]
for ax in bax.axs:
    ax.xaxis.set_major_locator(FixedLocator(grid_positions))  # Set the exact positions of the vertical grid lines
    ax.grid(True, which='major', axis='x', linestyle='--', alpha=1.0, linewidth=2.0)

# Add right and top borders
for ax in bax.axs:
    ax.spines['right'].set_visible(True)
    ax.spines['top'].set_visible(True)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.spines['bottom'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.tick_params(axis='x', width=4, colors='black', labelsize=30, labelcolor='black')
    ax.tick_params(axis='y', width=4, colors='black', labelsize=30, labelcolor='black')

# Only add horizontal lines on the top and right
bax.axs[1].spines['top'].set_visible(False)  # Hide the bottom border

# Show the plot
plt.show()
```

#### Example 4

```python
# Example 4
filename = 'Gene_Data/example_4.txt'
arrays, gene_dict = process_gene_data(filename)

# Generate the range of read lengths
readlengh_range_short =  np.linspace(300, 500, 10, endpoint=True, dtype=int)
readlengh_range_long =  np.linspace(1900, 2100, 10, endpoint=True, dtype=int)

# Initialize the condition number list
condition_number_high = []
condition_number_low = []

# Compute the condition number for each read length
for readlengh in readlengh_range_short:
    C = computeC(arrays, readlengh, gene_dict)
    eigenvalues, eigenvectors = np.linalg.eig(C.T @ C)
    condition_number_temp = max(eigenvalues) / min(eigenvalues)
    condition_number_high.append(condition_number_temp)
    
for readlengh in readlengh_range_long:
    C = computeC(arrays, readlengh, gene_dict)
    eigenvalues, eigenvectors = np.linalg.eig(C.T @ C)
    condition_number_temp = max(eigenvalues) / min(eigenvalues)
    condition_number_low.append(condition_number_temp)
```

```python
# Create the main figure
fig = plt.figure(figsize=(15, 12), dpi=600)

# Create a broken y-axis plot
bax = brokenaxes(xlims=((290, 500), (1900, 2110)), ylims=((0, 300), (400, 900)), hspace=.2, d=0.009, wspace=.15)  # Adjust ylims, hspace, and wspace to fit your data

# To make the image look better, scale the data when the condition number is low
condition_number_low = [i * 76 for i in condition_number_low]

# Plot the data
line1 = bax.plot(list(readlengh_range_short) + list(readlengh_range_long), condition_number_high + condition_number_low, 'o-', markersize=10, color='crimson', label='Condition Number', linewidth=4)

# Set axis labels
bax.set_xlabel('Read Length', fontsize=45, labelpad=45)
bax.set_ylabel('Condition Number', fontsize=45, labelpad=85)

# Set x-axis ticks and labels
xticks_short = [300, 400, 500]
xticks_long = [1950, 2025, 2100]
bax.axs[0].set_xticks(xticks_short)
bax.axs[0].set_xticklabels(xticks_short, fontsize=30, rotation=0, ha="center")
bax.axs[1].set_xticks(xticks_long)
bax.axs[1].set_xticklabels(xticks_long, fontsize=30, rotation=0, ha="center")

# Set y-axis ticks and labels
yticks = np.linspace(0, 300, 4)
yticklabels = np.linspace(0, 3, 4)
bax.axs[2].set_yticks(yticks)
bax.axs[2].set_yticklabels([f'{int(label)}' for label in yticklabels], fontsize=30, rotation=0, ha="center")
bax.axs[3].set_yticks(yticks)
bax.axs[3].set_yticklabels([f'{int(label)}' for label in yticklabels], fontsize=30, rotation=0, ha="center")

# Adjust the distance between y-axis tick labels and the y-axis
bax.axs[2].tick_params(axis='y', pad=15)
bax.axs[3].tick_params(axis='y', pad=15)

# Create proxy artist
proxy = [Line2D([0], [0], color='crimson', marker='o', linestyle='-', markersize=5)]

# Add grid lines
bax.grid(True, linestyle='--', alpha=1.0, linewidth=2.0)

# Add right and top borders
for ax in bax.axs:
    ax.spines['right'].set_visible(True)
    ax.spines['top'].set_visible(True)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.spines['bottom'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.tick_params(axis='x', width=4, colors='black', labelsize=30, labelcolor='black')
    ax.tick_params(axis='y', width=4, colors='black', labelsize=30, labelcolor='black')

# Only add horizontal lines on the top and right
bax.axs[1].spines['top'].set_visible(False)  # Hide the bottom border
bax.axs[0].spines['right'].set_visible(False)  # Hide the right border
bax.axs[2].spines['top'].set_visible(False)  # Hide the top border
bax.axs[3].spines['top'].set_visible(False)  # Hide the top border
bax.axs[1].spines['top'].set_visible(True)  # Show the top border

# Show the plot
plt.show()
```
